### Ce notebook a un rôle de vérification rapide des données de prix du Bitcoin téléchargées depuis Kaggle. 

### Il ne fait pas partie du pipeline de feature engineering ou de modélisation : il sert uniquement à valider que les données brutes sont correctement téléchargées et lisibles par Spark.


```python
from pathlib import Path
from pyspark.sql import SparkSession

NB_DIR = Path.cwd()
PROJECT_ROOT = NB_DIR.parent

prices_path = PROJECT_ROOT / "data" / "prices" / "btcusd_1-min_data.csv"

spark = SparkSession.builder.appName("bda-price-check").getOrCreate()
spark.sparkContext.setLogLevel("ERROR")
df = spark.read.option("header", True).csv(str(prices_path))
df.printSchema()
df.select("Timestamp","Open","High","Low","Close","Volume").show(5, truncate=False)
spark.stop()
```
